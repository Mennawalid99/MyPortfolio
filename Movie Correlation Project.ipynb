```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None



df = pd.read_csv(r'C:\Users\Menna\Downloads\movies.csv')
```

```python
df.head()
```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
df.dtypes
```

```python
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})')
```

```python
df.head()
```

```python
df = df.sort_values(by =['gross'],inplace=False,ascending= False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
#Comparing Budget Vs Gross

plt.scatter(x = df['budget'],y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Film Budget')
plt.show()
```

```python
df.head()
```

```python
#Comparing budget vs gross using seaborn
sns.regplot(x = 'budget',y = 'gross', data = df, scatter_kws ={"color":"blue"}, line_kws ={"color":"red"})
```

```python
df.corr(method = 'pearson')
```

```python
correlation_matrix = df.corr(method = 'pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df.apply(lambda x: x.factorize()[0]).corr(method='pearson').head()     
```

```python
correlation_matrix = df.apply(lambda x: x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
correlation_mat = df.apply(lambda x: x.factorize()[0]).corr()

corr_pairs = correlation_mat.unstack()

print(corr_pairs.head())
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr.head()
```
